```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('placement.csv')
df.shape
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
plt.scatter(df.cgpa , df.package)
plt.plot(X_train,lr.predict(X_train),color = "red")
plt.xlabel('CGPA')
plt.ylabel('Paclage (in lpa)')
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

to find the error or score we have to pass two parameters one is actual values and second is predicted values

```python
#predicted values
y_pred = lr.predict(X_test)
y_pred
```

```python
# actual values
y_test.values
```

```python
# MAE
print('MAE(in lpa):',mean_absolute_error(y_test,y_pred),'lack')
```

```python
#MSE
print("MSE",mean_squared_error(y_pred,y_test))
```

```python
#RMSE
print("RMSE(in lpa):",np.sqrt(mean_squared_error(y_pred,y_test)))
```

```python
# R2_score
print('R2_score(goodness of model):',r2_score(y_test,y_pred))
r2 = r2_score(y_test,y_pred)
```

```python
# Adjusted R2_score
# we have to find no of rows in X_test 
X_test.shape
```

```python
# Adjusted R2_score
1 - ((1-r2)*(40-1) / (40-1-1))
```
